```python
import pylab as plt
```

```python
mySample = []
myLinear = []
myQuadratic = []
myCubic = []
myExponential = []

for i in range(0, 30):
    mySample.append(i)
    myLinear.append(i)
    myQuadratic.append(i**2)
    myCubic.append(i**3)
    myExponential.append(1.5**i)
```

```python
plt.plot(mySample, myLinear)
plt.plot(mySample, myQuadratic)
plt.plot(mySample, myCubic)
plt.plot(mySample, myExponential)
```

```python
plt.figure('lineal')
plt.plot(mySample, myLinear)
plt.figure('Quadratic')
plt.plot(mySample, myQuadratic)
plt.figure('Cubic')
plt.plot(mySample, myCubic)
plt.figure('Exponential')
plt.plot(mySample, myExponential)
```

```python
plt.figure('lineal')
plt.xlabel('Sample points')
plt.ylabel('Linear function')
plt.plot(mySample, myLinear)
plt.figure('Quadratic')
plt.xlabel('Sample points')
plt.ylabel('Quadratic function')
plt.plot(mySample, myQuadratic)
plt.figure('Cubic')
plt.xlabel('Sample points')
plt.ylabel('Cubir function')
plt.plot(mySample, myCubic)
plt.figure('Exponential')
plt.xlabel('Sample points')
plt.ylabel('Exponential function')
plt.plot(mySample, myExponential)
```

```python
plt.figure('lineal')
plt.title('LINEAR')
plt.xlabel('Sample points')
plt.ylabel('Linear function')
plt.plot(mySample, myLinear)
plt.figure('Quadratic')
plt.title('QUADRATIC')
plt.xlabel('Sample points')
plt.ylabel('Quadratic function')
plt.plot(mySample, myQuadratic)
plt.figure('Cubic')
plt.title('CUBIC')
plt.xlabel('Sample points')
plt.ylabel('Cubir function')
plt.plot(mySample, myCubic)
plt.figure('Exponential')
plt.title('EXPONENTIAL')
plt.xlabel('Sample points')
plt.ylabel('Exponential function')
plt.plot(mySample, myExponential)
```

```python
plt.figure('lineal')
plt.clf()
plt.title('LINEAR')
plt.xlabel('Sample points')
plt.ylabel('Linear function')
plt.plot(mySample, myLinear)
plt.figure('Quadratic')
plt.clf()
plt.title('QUADRATIC')
plt.xlabel('Sample points')
plt.ylabel('Quadratic function')
plt.plot(mySample, myQuadratic)
plt.figure('Cubic')
plt.clf()
plt.title('CUBIC')
plt.xlabel('Sample points')
plt.ylabel('Cubir function')
plt.plot(mySample, myCubic)
plt.figure('Exponential')
plt.clf()
plt.title('EXPONENTIAL')
plt.xlabel('Sample points')
plt.ylabel('Exponential function')
plt.plot(mySample, myExponential)
```

```python
plt.figure('lineal')
plt.clf()
plt.ylim(0, 400)
plt.title('LINEAR')
plt.xlabel('Sample points')
plt.ylabel('Linear function')
plt.plot(mySample, myLinear)
plt.figure('Quadratic')
plt.clf()
plt.ylim(0, 400)
plt.title('QUADRATIC')
plt.xlabel('Sample points')
plt.ylabel('Quadratic function')
plt.plot(mySample, myQuadratic)
plt.figure('Cubic')
plt.clf()
plt.ylim(0, 100000)
plt.title('CUBIC')
plt.xlabel('Sample points')
plt.ylabel('Cubir function')
plt.plot(mySample, myCubic)
plt.figure('Exponential')
plt.clf()
plt.ylim(0, 100000)
plt.title('EXPONENTIAL')
plt.xlabel('Sample points')
plt.ylabel('Exponential function')
plt.plot(mySample, myExponential)
```

```python
plt.figure('lineal quadratica')
plt.clf()
plt.ylim(0, 400)
plt.title('LINEAR vs. QUADRATICA')
plt.xlabel('Sample points')
plt.plot(mySample, myLinear)
plt.plot(mySample, myQuadratic)
```

```python
plt.figure('Cubic Exponential')
plt.clf()
plt.ylim(0, 100000)
plt.title('CUBIC vs. EXPONENTIAL')
plt.xlabel('Sample points')
plt.plot(mySample, myCubic)
plt.plot(mySample, myExponential)
```

```python
plt.figure('lineal quadratica')
plt.clf()
plt.ylim(0, 400)
plt.title('LINEAR vs. QUADRATICA')
plt.xlabel('Sample points')
plt.plot(mySample, myLinear, label='lineal')
plt.plot(mySample, myQuadratic, label='quadratic')
plt.legend(loc='upper left')
```

```python
plt.figure('Cubic Exponential')
plt.clf()
plt.ylim(0, 100000)
plt.title('CUBIC vs. EXPONENTIAL')
plt.xlabel('Sample points')
plt.plot(mySample, myCubic, label='cubic')
plt.plot(mySample, myExponential, label='exponential')
plt.legend()
```

'b-', 'ro', 'g^', 'r--'. Estos smbolos hacen referencia al tipo de lnea del grfico.
La primera letra corresponde al color, el otro smbolo al tipo de grfico.
linewidth, representa el grosor de la lnea del grfico.

```python
plt.figure('lineal quadratica')
plt.clf()
plt.ylim(0, 400)
plt.title('LINEAR vs. QUADRATICA')
plt.xlabel('Sample points')
plt.plot(mySample, myLinear, 'b-', label='lineal', linewidth=2.0)
plt.plot(mySample, myQuadratic, 'ro', label='quadratic', linewidth=3.0)
plt.legend(loc='upper left')
```

```python
plt.figure('Cubic Exponential')
plt.clf()
plt.ylim(0, 100000)
plt.title('CUBIC vs. EXPONENTIAL')
plt.xlabel('Sample points')
plt.plot(mySample, myCubic, 'g^', label='cubic', linewidth=4.0)
plt.plot(mySample, myExponential, 'r--', label='exponential', linewidth=5.0)
plt.legend()
```

plt.subplot(211), esto es para organizar los grficos que estn en la figura. El 2 dice que son dos filas, el 1 que es una columna, y el 1 dice en qu lugar estoy.
plt.subplot(121), 1 es una fila, 2 son dos columnas.
En este caso, tengo que ir especificando las caractersticas de cada grfico.

```python
plt.figure('lineal quadratica')
plt.clf()
plt.subplot(211)
plt.ylim(0, 900)
plt.plot(mySample, myLinear, 'b-', label='lineal', linewidth=2.0)
plt.legend(loc='upper left')
plt.subplot(212)
plt.ylim(0, 900)
plt.plot(mySample, myQuadratic, 'ro', label='quadratic', linewidth=3.0)
plt.legend(loc='upper left')
plt.title('lineal vs. quadratic')
plt.xlabel('Sample points')
```

```python
plt.figure('Cubic Exponential')
plt.clf()
plt.subplot(121)
plt.ylim(0, 100000)
plt.plot(mySample, myCubic, 'g^', label='cubic', linewidth=4.0)
plt.legend(loc='upper left')
plt.xlabel('Sample points')
plt.subplot(122)
plt.ylim(0, 100000)
plt.plot(mySample, myExponential, 'r--', label='exponential', linewidth=5.0)
plt.legend(loc='upper left')
plt.title('cubic vs. exponential')
plt.xlabel('Sample points')
```

Puedo cambiar la escala, por ejemplo, cambiarla a logartmica.

```python
plt.figure('cube exp log')
plt.clf()
plt.plot(mySample, myCubic, 'g^', label='cubic', linewidth=4.0)
plt.plot(mySample, myExponential, 'r--', label='exponential', linewidth=5.0)
plt.yscale('log')
plt.legend()
plt.title('cubic vs exponential log')
```
